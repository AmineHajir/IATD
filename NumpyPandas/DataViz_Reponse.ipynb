# Plot Types

There are several plot types built into pandas; most of them are statistical by nature:

<pre>
df.plot.hist()     histogram
df.plot.bar()      bar chart
df.plot.barh()     horizontal bar chart
df.plot.line()     line chart
df.plot.area()     area chart
df.plot.scatter()  scatter plot
df.plot.box()      box plot
df.plot.kde()      kde plot
df.plot.hexbin()   hexagonal bin plot
df.plot.pie()      pie chart</pre>

You can also call specific plots by passing their name as an argument, as with `df.plot(kind='area')`.

Let's start going through them! First we'll look at the data:

```python
# Plot the histogram of df1 col A - 
#df1.hist(column='A')
df1['A'].plot.hist()
```

```python
# Insert lines between vertical bins - Bring the axis to the edge of the graph
# You can use any matplotlib color spec for edgecolor, such as 'b', 'g', 'r', 'c', 'm', 'y', 'k', 'w'
# or the string representatÂ° of a float value for shades of grey, such as '0.5'
# For autoscale the axis can be set to 'x', 'y' or 'both'
df1['A'].plot.hist(ec='black').autoscale(tight=True)
```

```python
#change the number of bins 
df1['A'].plot.hist(ec='k',bins=np.arange(-4,4,0.2)).autoscale(tight=True)
```

```python
#Barplots are similar to histograms, except that they deal with discrete data, and often reflect multiple variables - Show the barplot for df2
df2.plot.bar()
```

```python
#show the horizontal bar plot 
df2.plot.barh()
```

```python
# Use .bar() and show the stacked version of your bar plot
df2.plot.bar(stacked=True)
```

```python
#Line plots are used to compare two or more variables. By default the x-axis values are taken from the index - Show the lineplot for df2 col a
df2.plot.line(y='a', figsize=(12,3), lw=3)
```

```python
# Compair the col a, b, c from df2 with line plot
df2.plot.line(y=['a','b','c'],figsize=(12,4), lw=2)
```

```python
# Area plots represent cumulatively stacked line plots where the space between lines is emphasized with colors - Show the area plot for df2
df2.plot.area()
```

```python
#Scatter plots are a useful tool to quickly compare two variables, and to look for possible trends - Show the scatter plot from df1 for col A and B
df1.plot.scatter(x='A',y='B')
```

```python
#show the same scatter plot plus colormaps
df1.plot.scatter(x='A', y='B', c='C', cmap='Paired') 
```

```python
#show the same box plot for df2
df2.boxplot()
```

```python
# Show the boxplot for col b and e grouped by col e
df2[['e','a']].boxplot(by='e')
```

```python
#use kernel density estimation plot and show the densitity of column a in df2
df2['a'].plot.kde()
```

```python
#show the density of all columns with kde plot for df2
df2.plot.kde()
```

```python
#use the below line of code and show the hexagonal bin plot for the generated number
#hexagonal bin plot is useful for Bivariate Data, alternative to scatterplot
# FIRST CREATE A DATAFRAME OF RANDOM VALUES
df = pd.DataFrame(np.random.randn(1000, 2), columns=['a', 'b'])
```

```python
# MAKE A HEXAGONAL BIN PLOT
df.plot.hexbin(x='a',y='b',gridsize=20,cmap='Oranges')
```

```python
#set color of the plot to red, with dashed/dotted style
df2['c'].plot.line(figsize=(8,3),c='red',ls='-.',lw=2)
```

```python
#show the line plot for entitre col in df2 
df2.plot(figsize=(10,4))
```

```python
# Add title, x-lable, y-lable to the plot - Set the legend to the lower right of the plot
ax = df2.plot.line(figsize=(10,5), title="Column Dataviz")
ax.set(xlabel="Abscisse", ylabel="OrdonnÃ©e")
ax.autoscale(axis='x', tight=True)
ax.legend(loc=4)

# Place Legend in the lower left
ax.legend(loc=3)

# Move the legend a little to the right and up 
ax.legend(bbox_to_anchor=(0.1,0.2))

# place the legend outside the plot
ax.legend(bbox_to_anchor=(1,1))
```

For starters we can pass a location code. <tt>.legend(loc=1)</tt> places the legend in the upper-right corner of the plot.<br>Alternatively we can pass a location string: <tt>.legend(loc='upper right')</tt> does the same thing.

```python
#import pandas and read df3.csv
path3 = "/content/gdrive/MyDrive/Numpy Pandas/DataViz_df3.csv"
df3 = pd.read_csv(path3)
df3
```

```python
#print len and head of the csv file
print(df3.shape)
df3.head()
```

```python
#1. create scatter plot of 'produced' vs 'defective'
df3.plot.scatter(x='produced', y='defective', figsize=(10,5))
```

```python
#2. Create a histogram of the 'produced' column - Tightening the x-axis and adding lines between bars
df3["produced"].plot.hist(ec='k', bins=np.arange(45,110,5)).autoscale(tight=True, axis='x') 
```

```python
#4. Create a boxplot that shows 'produced' for each 'weekday' (hint: this is a groupby operation)
df3[['weekday','produced']].boxplot(by='weekday') 
```

```python
#5. Create a KDE plot of the 'defective' column - Increase the linewidth and make the linestyle dashed
df3['defective'].plot.kde(ls=':',lw=4, figsize=(10,5))
```

```python
#7. Create a blended area plot of all the columns for just the rows up to 30. (hint: use .loc)
df3[:31].plot.area(stacked=False, alpha=0.6);
```

```python
#Notice how the legend in our previous figure overlapped some of actual diagram - Can you figure out how to display the legend outside of the plot?
df3[:31].plot.area(stacked=False, alpha=0.6).legend(bbox_to_anchor=(1,1));
```
